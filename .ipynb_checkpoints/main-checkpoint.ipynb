```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_path = "covid_death.csv"
data = pd.read_csv(dataset_path)
```

```python
print("Initial Dataset Preview:\n", data.head())
print("\nDataset Info:")
data.info()
```

```python
# Handling missing values
data = data.dropna()  # Drop rows with missing values
```

```python
# Convert date columns to datetime format
data['start_date'] = pd.to_datetime(data['start_date'])
data['end_date'] = pd.to_datetime(data['end_date'])
```

```python
#Line chart of COVID-19 deaths over time for a United States
country_filter = "United States"
country_data = data[data['country'] == country_filter]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(country_data['start_date'], country_data['covid_deaths'], marker='o', color='skyblue')
plt.title(f"COVID-19 Deaths Over Time in {country_filter}")
plt.xlabel("Start Date")
plt.ylabel("COVID-19 Deaths")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
#Bar chart of excess deaths per region
plt.figure(figsize=(12, 6))
region_excess_deaths = data.groupby('region')['excess_deaths'].sum().sort_values(ascending=False)
region_excess_deaths.plot(kind='bar', color='salmon', edgecolor='black')
plt.title("Excess Deaths by Region")
plt.xlabel("Region")
plt.ylabel("Excess Deaths")
plt.xticks(rotation=45)
plt.show()
```

```python
#Scatter plot of COVID deaths vs. excess deaths per 100k population
plt.figure(figsize=(8, 6))
sns.scatterplot(
    x=data['covid_deaths_per_100k'],
    y=data['excess_deaths_per_100k'],
    hue=data['region'],
    palette='viridis',
    s=100
)
plt.title("COVID Deaths vs Excess Deaths per 100k Population")
plt.xlabel("COVID Deaths per 100k")
plt.ylabel("Excess Deaths per 100k")
plt.legend(title="Region", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid()
plt.show()
```

```python
data.to_csv("cleaned_covid_death.csv", index=False)
print("Data cleaning and visualization completed. Cleaned data saved as 'cleaned_covid_death.csv'.")
```

```python
# Load the cleaned dataset
data = pd.read_csv('cleaned_covid_death.csv')
```

```python
# Group Data by Year
data['year'] = pd.to_datetime(data['start_date']).dt.year
yearly_data = data.groupby('year')[['covid_deaths', 'excess_deaths']].sum()
```

```python
# Plot yearly trends
plt.figure(figsize=(10, 6))
yearly_data.plot(kind='bar', stacked=True, color=['skyblue', 'salmon'], figsize=(10, 6))
plt.title("Yearly COVID-19 Deaths and Excess Deaths")
plt.xlabel("Year")
plt.ylabel("Number of Deaths")
plt.legend(["COVID-19 Deaths", "Excess Deaths"])
plt.grid()
plt.show()
```

```python
# Calculate Percent Contribution of COVID-19 Deaths
data['covid_death_pct'] = (data['covid_deaths'] / data['total_deaths']) * 100
percent_contribution = data.groupby('region')['covid_death_pct'].mean().sort_values(ascending=False)
```

```python
# Plot the percentage contribution
percent_contribution.plot(kind='bar', color='teal', figsize=(12, 6), edgecolor='black')
plt.title("Average COVID-19 Deaths as Percentage of Total Deaths by Region")
plt.xlabel("Region")
plt.ylabel("Percentage (%)")
plt.grid()
plt.show()
```

```python
# Identify Regions with the Highest and Lowest Impact
region_ranking = data.groupby('region')[['excess_deaths', 'covid_deaths']].sum().sort_values(by='excess_deaths', ascending=False)
print("Regions with the Highest Excess Deaths:")
print(region_ranking.head(5))
print("\nRegions with the Lowest Excess Deaths:")
print(region_ranking.tail(5))
```

```python
# Growth Rate of COVID-19 Deaths
data['growth_rate'] = data['covid_deaths'].pct_change() * 100
```

```python
# Plot growth rate
plt.figure(figsize=(10, 6))
plt.plot(data['start_date'], data['growth_rate'], marker='o', linestyle='-', color='green')
plt.title("Growth Rate of COVID-19 Deaths Over Time")
plt.xlabel("Date")
plt.ylabel("Growth Rate (%)")
plt.grid()
plt.show()
```
